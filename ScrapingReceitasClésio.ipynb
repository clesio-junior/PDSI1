A primeira coisa a fazer no trabalho  verificar se est retornando o cdigo 200.

Formatar a linguagem html corretamente usando html.parser.

```python
from bs4 import BeautifulSoup
if(resposta.status_code==200):
  soup = BeautifulSoup(resposta.content, 'html.parser')
  print(soup.prettify())
```

# **Encontrando os elementos interessantes na pgina, com o find.**

Foram escolhidas as tags "`h3`" para os ttulos das receitas e "`a href`" para os links.

```python
#print(soup.title)

div_tags = soup.find_all('div', class_="info bgbege")

for tag in div_tags:
    h3_tag = tag.find('h3')
    if h3_tag:
        print("Ttulo:", h3_tag.text)

    a_tag = tag.find('a')
    if a_tag and 'href' in a_tag.attrs:
        print("Link:", a_tag['href'])
```

Pegando a ocorrncia especifica do "Suco de Luz", acessando a quinta ocorrncia retornada pela busca armazenada em `div_tags` na posio `[4]`.

Pego apenas o link dessa receita especfica e armazeno na varivel **`a`**.

```python
#a_tag = div_tags[4].find('a')
#print(a_tag['href'])

a = div_tags[4].find('a')['href']

print(a)
```

Fazendo o webscraping do suco escolhido, com o link armazenado na varivel "`a`":

```python
respostaSuco = requests.get(a)

print("URL = ",respostaSuco.url)

print("Status = ",respostaSuco.status_code)

print("Contedo:", respostaSuco.content)
```

Novamente, se retornar cdigo `200`, fao o `parser.html` para organizar visualmente o html relacionado  receita.

```python
if(resposta.status_code==200):
  soup = BeautifulSoup(respostaSuco.content, 'html.parser')
  print(soup.prettify())
```


# h1 suco de luz

```python
nomeReceita = soup.h1.get_text()

print("Ttulo: ",nomeReceita)
```

```python
#Executando o print find_all('ul') pra saber qual das ocorrencias  a que preciso:
ul_tags = soup.find_all('ul')
#for topic in ul_tags:
#  print(topic.text)

#Descobri que preciso da terceira ocorrncia:
print("\n\nAQUI = ", ul_tags[2])
```

Pegando cada um dos ingredientes:


```python
li_tags = ul_tags[2].find_all('li')

for item in li_tags:
  print(item.get_text())
```

Adicionando os ingredientes numa lista:

```python
listaCompras = []

for item in li_tags:
  textoItem = item.get_text()

  #remove o \n do incio das frases
  if('\n' in textoItem):
    textoItem = textoItem.replace('\n', "")

  listaCompras.append(textoItem)
  print(item.get_text())


print("\nLista de compras: ", listaCompras)
```

# Agora, para o '**passo a passo**':

```python
#Executando o print find_all('ol') pra saber qual das ocorrencias  a que preciso:
ol_tags = soup.find_all('ol')
#for topic in ul_tags:
#  print(topic.text)

#Descobri que preciso da primeira ocorrncia:
print("\n\nAQUI = ", ol_tags[0])
```

```python

li_tags1 = ol_tags[0].find_all('li')

#criando lista pra salvar cada uma das etapas
passos = []

for item in li_tags1:
  textoItem = item.get_text()

  #remove o \n do incio das frases
  if('\n' in textoItem):
    textoItem = textoItem.replace('\n', "")

  passos.append(textoItem)
  print(item.get_text())


print("\n\n\tLista de passo a passo: ", passos)
```

# PEGAR LINK DA IMAGEM

# **Imprimindo todos os dados extraidos**

```python
print("\t\t{}\n".format(nomeReceita))
print("Link da imagem: ", imagem['src'])

print("\n\t Ingredientes:\n")
for item in listaCompras:
  print(item)



print("\n\tModo de Preparo:\n")
i = 1
for passo in passos:
  print("{} passo: {}". format(i, passo))
  i = i + 1

```

# GERAR EXCEL

```python
from openpyxl import Workbook
from openpyxl.styles import PatternFill, Alignment

# Define as cores de preenchimento
orangeFill = PatternFill(start_color='FFC000',
                           end_color='FFC000',
                           fill_type='solid')
blueFill = PatternFill(start_color='00B0F0',
                         end_color='00B0F0',
                         fill_type='solid')

# Cria um novo Workbook e remove a planilha padro
wb = Workbook()
del wb['Sheet']

# Adiciona uma nova planilha
sheet = wb.create_sheet('Receita Suco de Luz')

# Ttulos das colunas
sheet.cell(1, 1).value = "Ingredientes"
sheet.cell(1, 1).fill = orangeFill
sheet.cell(1, 1).alignment = Alignment(horizontal='center', vertical='center')

sheet.cell(1, 2).value = "Modo de Preparo"
sheet.cell(1, 2).fill = blueFill
sheet.cell(1, 2).alignment = Alignment(horizontal='center', vertical='center')


# Adiciona os ingredientes na coluna A
for i, ingrediente in enumerate(listaCompras):
    sheet.cell(i+2, 1).value = ingrediente

# Adiciona o modo de preparo na coluna B
for i, passo in enumerate(passos):
    sheet.cell(i+2, 2).value = passo

# Salva o arquivo
wb.save("Receita_Suco_de_Luz_teste2.xlsx")
```
